```python
from soyspacing.countbase import CountSpace

model_fname = "./space-correct.model"
model = CountSpace()
model.load_model(model_fname, json_format=False) #띄어쓰기 모델 학습 불러오기
```

```python
import pandas as pd

df = pd.read_excel('교통사고처리특례법_교정.xlsx')

for i in range(len(df)):
    
    if pd.isnull(df['판시사항'][i]) is False: #결측 확인 (처리안해주면 float오류발생)
        sent_corrected1, tags = model.correct(df['판시사항'][i]) #단어간 띄어쓰기 교정
        df['판시사항'][i]=sent_corrected1     
        
    if pd.isnull(df['판결요지'][i]) is False:
        sent_corrected2, tags = model.correct(df['판결요지'][i])
        df['판결요지'][i]=sent_corrected2
    
    if pd.isnull(df['참조조문'][i]) is False:
        sent_corrected3, tags = model.correct(df['참조조문'][i])
        df['참조조문'][i]=sent_corrected3
    
    if pd.isnull(df['참조판례'][i]) is False:
        sent_corrected4, tags = model.correct(df['참조판례'][i])
        df['참조판례'][i]=sent_corrected4

    
df.to_excel('교통사고처리특례법_교정.xlsx')
```

```python
import pandas as pd

#앞에 [1]없는 컬럼들 추가해주는 파일

df = pd.read_excel('교통사고처리특례법_번호없는거교정.xlsx')

for i in range(len(df)):
    if pd.isnull(df['판시사항'][i]) is False:
        if df['판시사항'][i][0]!='[':
            df['판시사항'][i]='[1]'+df['판시사항'][i]
        
    if pd.isnull(df['판결요지'][i]) is False:
        if df['판결요지'][i][0]!='[':
            df['판결요지'][i]='[1]'+df['판결요지'][i]
    
    if pd.isnull(df['참조조문'][i]) is False:
        if df['참조조문'][i][0]!='[':
            df['참조조문'][i]='[1]'+df['참조조문'][i]
     
    if pd.isnull(df['참조판례'][i]) is False:
        if df['참조판례'][i][0]!='[':
            df['참조판례'][i]='[1]'+df['참조판례'][i]
        

df.to_excel('교통사고처리특례법_번호없는거교정.xlsx')
```
